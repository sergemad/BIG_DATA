# Visualisation

### Highest PTS per season

```python
df_Steph_stats['Max_PTS'] = df_Steph_stats.groupby(['Season_year'])['PTS'].transform(max)
#df_Steph_stats.head()
```

```python
plt.subplots(figsize=(15, 6))
plt.bar(df_Steph_stats.groupby(['Season_year'])['Max_PTS'].mean().index,
        df_Steph_stats.groupby(['Season_year'])['Max_PTS'].mean().values,
        width = 0.6, color = 'yellow', edgecolor = 'blue', linewidth = 2)
plt.title('Highest points per season')
plt.show()
```

### Average PTS per season

```python
df_Steph_stats['S_Mean_PTS'] = df_Steph_stats.groupby(['Season_year'])['PTS'].transform(np.mean)
df_Steph_stats.head()
```

```python
plt.subplots(figsize=(15, 6))
plt.bar(df_Steph_stats.groupby(['Season_year'])['S_Mean_PTS'].mean().index,
        df_Steph_stats.groupby(['Season_year'])['S_Mean_PTS'].mean().values,
        width = 0.6, color = 'yellow', edgecolor = 'blue', linewidth = 2)
plt.title('Average points per season')
plt.show()
```

### Average and highest points per opposition

```python
df_Steph_stats['OPP_Mean_PTS'] = df_Steph_stats.groupby(['OPP'])['PTS'].transform(np.mean)
df_Steph_stats.head()
```

```python
df_Steph_stats['OPP_Max_PTS'] = df_Steph_stats.groupby(['OPP'])['PTS'].transform(max)
df_Steph_stats.head()
```

```python
df_Steph_stats['OPP_Min_PTS'] = df_Steph_stats.groupby(['OPP'])['PTS'].transform(min)
df_Steph_stats.head()
```

```python
plt.subplots(figsize=(15, 6))
plt.bar(df_Steph_stats.groupby(['OPP'])['OPP_Mean_PTS'].mean().index,
        df_Steph_stats.groupby(['OPP'])['OPP_Mean_PTS'].mean().values,
        width = 0.6, color = 'yellow', edgecolor = 'blue', linewidth = 2)
plt.title('Average points per opposition')
plt.show()
```

```python
plt.subplots(figsize=(15, 6))
plt.bar(df_Steph_stats.groupby(['OPP'])['OPP_Max_PTS'].mean().index,
        df_Steph_stats.groupby(['OPP'])['OPP_Max_PTS'].mean().values,
        width = 0.6, color = 'yellow', edgecolor = 'blue', linewidth = 2)
plt.title('Highest points per opposition')
plt.show()
```

```python
plt.subplots(figsize=(15, 6))
plt.bar(df_Steph_stats.groupby(['OPP'])['OPP_Min_PTS'].mean().index,
        df_Steph_stats.groupby(['OPP'])['OPP_Min_PTS'].mean().values,
        width = 0.6, color = 'yellow', edgecolor = 'blue', linewidth = 2)
plt.title('minimum points per opposition')
plt.show()
```
